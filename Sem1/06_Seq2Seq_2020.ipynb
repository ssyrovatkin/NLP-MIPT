```
import random

SOURCE_FILE = "data/ru.txt"
TARGET_FILE = "data/en.txt"
TRAIN_SOURCE_FILE = "data/ru_train.txt"
TRAIN_TARGET_FILE = "data/en_train.txt"
VAL_SOURCE_FILE = "data/ru_val.txt"
VAL_TARGET_FILE = "data/en_val.txt"
TEST_SOURCE_FILE = "data/ru_test.txt"
TEST_TARGET_FILE = "data/en_test.txt"

train_pairs = []
val_pairs = []
test_pairs = []
random.seed(42)
with open(SOURCE_FILE, "r") as source_file, open(TARGET_FILE, "r") as target_file:
    for source, target in zip(source_file, target_file):
        pairs = None
        random_val = random.random() 
        if random_val < 0.95:
            pairs = train_pairs
        elif random_val < 0.975:
            pairs = val_pairs
        else:
            pairs = test_pairs
        pairs.append((source.strip(), target.strip()))

random.shuffle(train_pairs)
random.shuffle(val_pairs)
random.shuffle(test_pairs)

with open(TRAIN_SOURCE_FILE, "w") as source_file, open(TRAIN_TARGET_FILE, "w") as target_file:
    for source, target in train_pairs:
        source_file.write(source + "\n")
        target_file.write(target + "\n")

with open(VAL_SOURCE_FILE, "w") as source_file, open(VAL_TARGET_FILE, "w") as target_file:
    for source, target in val_pairs:
        source_file.write(source + "\n")
        target_file.write(target + "\n")

with open(TEST_SOURCE_FILE, "w") as source_file, open(TEST_TARGET_FILE, "w") as target_file:
    for source, target in test_pairs:
        source_file.write(source + "\n")
        target_file.write(target + "\n")
```

```
from pyonmttok import Tokenizer

pyonmt_tokenizer = Tokenizer("conservative", joiner_annotate=False)

def tokenize(text, lower=True):
    text = str(text).strip().replace("\n", " ").replace("\xa0", " ")
    if lower:
        text = text.lower()
    tokens, _ = pyonmt_tokenizer.tokenize(text)
    return tokens
```

## Трансформеры из коробки

```
from tqdm import tqdm
from nltk.translate.bleu_score import corpus_bleu

def evaluate_bleu(dataset, model, tokenizer, translate, nrows=None):
    assert translate is not None
    references = []
    hypotheses = []
    n = len(dataset) if nrows is None else min(len(dataset), nrows)
    for i in tqdm(range(n)):
        source, target = dataset.pairs[i]
        prediction = translate(source, model, tokenizer)
        references.append([" ".join(tokenize(target))])
        hypotheses.append(" ".join(tokenize(prediction)))
    return corpus_bleu(references, hypotheses)
```

```
import torch
from torch.utils.data import Dataset

class TranslationDataset(Dataset):
    def __init__(
        self,
        source_file,
        target_file,
        tokenizer,
        source_max_tokens_count=30,
        target_max_tokens_count=30
    ):
        self.pairs = []
        with open(source_file, "r") as source_file, open(target_file, "r") as target_file:
            for source, target in zip(source_file, target_file):
                self.pairs.append((source, target))
        self.tokenizer = tokenizer
        self.source_max_tokens_count = source_max_tokens_count
        self.target_max_tokens_count = target_max_tokens_count
    
    def __len__(self):
        return len(self.pairs)

    def __getitem__(self, index):
        source, target = self.pairs[index]
        inputs = self.tokenizer(
            source,
            add_special_tokens=True,
            max_length=self.source_max_tokens_count,
            padding="max_length",
            truncation=True
        )
        outputs = self.tokenizer(
            target,
            add_special_tokens=True,
            max_length=self.source_max_tokens_count,
            padding="max_length",
            truncation=True
        )
        decoder_input_ids = torch.tensor(outputs["input_ids"])
        decoder_attention_mask = torch.tensor(outputs["attention_mask"])
        labels = decoder_input_ids.clone()
        labels[decoder_attention_mask == 0] = -100
        return {
            "input_ids": torch.tensor(inputs["input_ids"]),
            "attention_mask": torch.tensor(inputs["attention_mask"]),
            "decoder_input_ids": decoder_input_ids,
            "decoder_attention_mask": decoder_attention_mask,
            "labels": labels
        }
```

### MBart

https://arxiv.org/abs/2001.08210

Seq2seq трансформеры специально для перевода. Предобучались на исправление шума на одноязычных корпусах сразу всей моделью, а не энкодером/декодором в отдельности.

Аналог от Гугла: mT5 https://arxiv.org/abs/2010.11934

```
import torch
from torch.utils.data import Dataset

class MBartTranslationDataset(Dataset):
    def __init__(
        self,
        source_file,
        target_file,
        tokenizer,
        max_tokens_count=30
    ):
        self.pairs = []
        with open(source_file, "r") as source_file, open(target_file, "r") as target_file:
            for source, target in zip(source_file, target_file):
                self.pairs.append((source, target))
        self.tokenizer = tokenizer
        self.max_tokens_count = max_tokens_count
    
    def __len__(self):
        return len(self.pairs)

    def __getitem__(self, index):
        source, target = self.pairs[index]
        batch = self.tokenizer.prepare_seq2seq_batch(
            source,
            src_lang="ru_RU",
            tgt_lang="en_XX",
            tgt_texts=target,
            return_tensors="pt",
            padding="max_length",
            truncation=True,
            max_length=self.max_tokens_count)
        return {
            "input_ids": batch["input_ids"][0],
            "attention_mask": batch["attention_mask"][0],
            "labels": batch["labels"][0]
        }
```

## Ванильный Torch и Seq2Seq с attention'ом

```
from collections import Counter
from typing import List, Tuple
import os
import torch

class Vocabulary:
    def __init__(self):
        self.index2word = list()
        self.word2index = dict()
        self.word2count = Counter()
        self.reset()

    def get_pad(self):
        return self.word2index["<pad>"]

    def get_bos(self):
        return self.word2index["<bos>"]

    def get_eos(self):
        return self.word2index["<eos>"]

    def get_unk(self):
        return self.word2index["<unk>"]

    def add_sentence(self, sentence):
        for word in tokenize(sentence.strip()):
            self.add_word(word)

    def add_word(self, word):
        if word not in self.word2index:
            self.word2index[word] = len(self.index2word)
            self.word2count[word] += 1
            self.index2word.append(word)
        else:
            self.word2count[word] += 1
    
    def has_word(self, word) -> bool:
        return word in self.word2index

    def add_file(self, filename: str):
        with open(filename, "r", encoding="utf-8") as r:
            for line in r:
                self.add_sentence(line)

    def get_index(self, word):
        if word in self.word2index:
            return self.word2index[word]
        return self.get_unk()

    def get_word(self, index):
        return self.index2word[index]

    def size(self):
        return len(self.index2word)

    def is_empty(self):
        empty_size = 4
        return self.size() <= empty_size

    def shrink(self, n):
        best_words = self.word2count.most_common(n)
        self.reset()
        for word, count in best_words:
            self.add_word(word)
            self.word2count[word] = count

    def reset(self):
        self.word2count = Counter()
        self.index2word = ["<pad>", "<bos>", "<eos>", "<unk>"]
        self.word2index = {word: index for index, word in enumerate(self.index2word)}

    def tokenize(self, sentence: str, max_length: int) -> List[int]:
        indices = [self.get_index(word) for word in tokenize(sentence)]
        indices = indices[:max_length-2]
        indices.insert(0, self.get_bos())
        indices.append(self.get_eos())
        mask = [1 for _ in range(len(indices))]
        indices += [self.get_pad() for _ in range(max_length - len(indices))]
        mask += [0 for _ in range(max_length - len(indices))]
        return torch.tensor(indices), torch.tensor(mask)
```

```
import numpy as np

import torch
import torch.nn as nn
import torch.nn.functional as F

class EncoderRNN(nn.Module):
    def __init__(self, input_size, embedding_dim, hidden_size, n_layers=3, dropout=0.3, bidirectional=True):
        super(EncoderRNN, self).__init__()

        num_directions = 2 if bidirectional else 1
        assert hidden_size % num_directions == 0
        hidden_size = hidden_size // num_directions

        self.embedding_dim = embedding_dim
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.n_layers = n_layers
        self.dropout = dropout
        self.bidirectional = bidirectional

        self.embedding = nn.Embedding(input_size, embedding_dim)
        self.rnn = nn.LSTM(embedding_dim, hidden_size, n_layers, dropout=dropout,
                           bidirectional=bidirectional, batch_first=True)

    def forward(self, source_ids, hidden=None):
        embedded = self.embedding(source_ids)
        outputs, hidden = self.rnn(embedded, hidden)

        if self.bidirectional:
            n = hidden[0].size(0)
            hidden = (torch.cat([hidden[0][0:n:2], hidden[0][1:n:2]], 2),
                      torch.cat([hidden[1][0:n:2], hidden[1][1:n:2]], 2))
        return outputs, hidden

class Generator(nn.Module):
    def __init__(self, hidden_size, output_size):
        super(Generator, self).__init__()

        self.hidden_size = hidden_size
        self.output_size = output_size

        self.out = nn.Linear(hidden_size, output_size)
        self.sm = nn.LogSoftmax(dim=1)

    def forward(self, inputs):
        assert inputs.size(1) == self.hidden_size
        return self.sm(self.out(inputs))
```

```
from torch.autograd import Variable


class DecoderRNN(nn.Module):
    def __init__(self, embedding_dim, hidden_size, output_size, max_length, n_layers=3, 
                 dropout=0.3, use_attention=True):
        super(DecoderRNN, self).__init__()

        self.embedding_dim = embedding_dim
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.n_layers = n_layers
        self.dropout = dropout
        self.max_length = max_length
        self.use_attention = use_attention

        self.embedding = nn.Embedding(output_size, embedding_dim)
        
        if self.use_attention:
            self.attn = nn.Linear(embedding_dim, hidden_size)
            self.attn_sm = nn.Softmax(dim=1)
            self.attn_ctx = nn.Linear(hidden_size + hidden_size, hidden_size)
            self.attn_tanh = nn.Tanh()
        
        self.rnn = nn.LSTM(embedding_dim, hidden_size, n_layers,
                           dropout=dropout, batch_first=True)
        self.generator = Generator(hidden_size, output_size)
        
    def step(self, batch_input, hidden, encoder_output):
        # hidden: (n_layers x B x N, n_layers x B x N)
        # encoder_output: B x L x N
        # h_t: B x N x 1
        # embedded:  B x E
        # attn_weights: B x 1 x L
        # context: B x 1 x N
        # rnn_input: B x N
        
        embedded = self.embedding(batch_input)
        _, hidden = self.rnn(embedded.unsqueeze(1), hidden)
        if self.use_attention:
            mapped_encoder_outputs = self.attn(encoder_output)
            h_t = hidden[0][-1].unsqueeze(2)
            attn_weights = self.attn_sm(torch.bmm(mapped_encoder_outputs, h_t)).transpose(1, 2)
            max_length = encoder_output.size(1)
            context = torch.bmm(attn_weights[:, :, :max_length], encoder_output)
            output = self.attn_tanh(self.attn_ctx(torch.cat((context, h_t.transpose(1, 2)), 2)))
        return output, hidden

    def init_state(self, batch_size, bos_index):
        initial_input = Variable(torch.zeros((batch_size,)).type(torch.LongTensor), requires_grad=False)
        initial_input = torch.add(initial_input, bos_index)
        initial_input = initial_input.cuda() if next(self.parameters()).is_cuda else initial_input
        return initial_input

    def forward(self, current_input, hidden, length, encoder_output, target_ids=None):
        batch_size = current_input.size(0)
        outputs = Variable(torch.zeros(batch_size, length, self.output_size), requires_grad=False)
        outputs = outputs.cuda() if next(self.parameters()).is_cuda else outputs

        for t in range(length):
            output, hidden = self.step(current_input, hidden, encoder_output)
            scores = self.generator.forward(output.squeeze(1))
            outputs[:, t, :] = scores
            if target_ids is None:
                top_indices = scores.topk(1, dim=1)[1].view(-1)
                current_input = top_indices
            else:
                current_input = target_ids[:, t]
        return outputs, hidden
```

```
from torch.optim import Adam
from pytorch_lightning import LightningModule

class Seq2Seq(LightningModule):
    def __init__(self, input_vocab_size, output_vocab_size,
                 embedding_dim=100, rnn_size=100,
                 encoder_n_layers=2, decoder_n_layers=2,
                 dropout=0.3, max_length=20,
                 bidirectional=True, use_attention=True):
        super(Seq2Seq, self).__init__()

        self.embedding_dim = embedding_dim
        self.input_vocab_size = input_vocab_size
        self.output_vocab_size = output_vocab_size
        self.rnn_size = rnn_size
        self.encoder_n_layers = encoder_n_layers
        self.decoder_n_layers = decoder_n_layers
        self.dropout = dropout
        self.max_length = max_length
        self.bidirectional = bidirectional
        self.use_attention = use_attention
        self.criterion = nn.NLLLoss()

        self.encoder = EncoderRNN(self.input_vocab_size, embedding_dim, rnn_size, dropout=dropout,
                                  n_layers=encoder_n_layers, bidirectional=bidirectional)
        self.decoder = DecoderRNN(embedding_dim, rnn_size, self.output_vocab_size, dropout=dropout,
                                  max_length=max_length, n_layers=decoder_n_layers, 
                                  use_attention=use_attention)

    def forward(self, source_ids, target_ids=None, bos_index=1):
        batch_size = source_ids.size(0)
        encoder_output, encoder_hidden = self.encoder.forward(source_ids)
        current_input = self.decoder.init_state(batch_size, bos_index)
        scores, _ = self.decoder.forward(
            current_input, encoder_hidden, self.max_length,
            encoder_output, target_ids
        )
        loss = None
        if target_ids is not None:
            loss = self.criterion(scores.transpose(1, 2), target_ids)
        return loss, scores
    
    def training_step(self, batch, batch_nb):
        train_loss, _ = self(**batch)
        self.log("train_loss", train_loss, prog_bar=True)
        return train_loss

    def validation_step(self, batch, batch_nb):
        val_loss, _ = self(**batch)
        self.log("val_loss", val_loss, prog_bar=True, logger=True)
    
    def configure_optimizers(self):
        optimizer = Adam(self.parameters(), lr=0.001)
        return [optimizer]
```

```
class Translator:
    def __init__(
        self,
        model: Seq2Seq,
        src_vocabulary: Vocabulary,
        tgt_vocabulary: Vocabulary, 
        use_cuda: bool
    ):
        self.model = model  # type: Seq2Seq
        self.src_vocabulary = src_vocabulary  # type: Vocabulary
        self.tgt_vocabulary = tgt_vocabulary  # type: Vocabulary
        self.use_cuda = use_cuda

    def translate_sentence(self, sentence: str):
        source_ids = self.sentence_to_tensor(sentence, self.src_vocabulary)
        outputs = self.translate(source_ids)

        translated = list(outputs.cpu().detach().numpy())
        words = []
        for i in translated:
            word = self.tgt_vocabulary.get_word(i)
            if word == "<bos>":
                continue
            if word == "<eos>" or word == "<pad>":
                break
            words.append(word)
        return " ".join(words)

    def translate(self, inputs):
        self.model.eval()
        inputs = inputs.to("cuda")
        batch_size = inputs.size(0)
        bos_index = self.tgt_vocabulary.get_bos()
        _, scores = self.model.forward(inputs, bos_index=bos_index)
        max_length = scores.size(1)
      
        output_variable = Variable(torch.zeros(batch_size, max_length).type(torch.LongTensor))
        output_variable = output_variable.cuda() if self.use_cuda else output_variable
        for t in range(max_length):
            output_variable[:, t] = scores[:, t, :].topk(1, dim=1)[1].view(-1)
        output_variable = output_variable.detach()
        return output_variable[0]

    def sentence_to_tensor(self, sentence, vocabulary):
        indices, _ = vocabulary.tokenize(sentence, max_length=self.model.max_length)
        tensor = indices.unsqueeze(0)
        tensor = tensor.cuda() if self.use_cuda else tensor
        return tensor
```

```
import numpy as np

import torch
import torch.nn as nn
import torch.nn.functional as F

class EncoderRNN(nn.Module):
    def __init__(self, input_size, embedding_dim, hidden_size, n_layers=3, dropout=0.3, bidirectional=True):
        super(EncoderRNN, self).__init__()

        num_directions = 2 if bidirectional else 1
        assert hidden_size % num_directions == 0
        hidden_size = hidden_size // num_directions

        self.embedding_dim = embedding_dim
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.n_layers = n_layers
        self.dropout = dropout
        self.bidirectional = bidirectional

        self.embedding = nn.Embedding(input_size, embedding_dim)
        self.rnn = nn.LSTM(embedding_dim, hidden_size, n_layers, dropout=dropout,
                           bidirectional=bidirectional, batch_first=True)

    def forward(self, source_ids, hidden=None):
        embedded = self.embedding(source_ids)
        outputs, hidden = self.rnn(embedded, hidden)

        if self.bidirectional:
            n = hidden[0].size(0)
            hidden = (torch.cat([hidden[0][0:n:2], hidden[0][1:n:2]], 2),
                      torch.cat([hidden[1][0:n:2], hidden[1][1:n:2]], 2))
        return outputs, hidden

class Generator(nn.Module):
    def __init__(self, hidden_size, output_size):
        super(Generator, self).__init__()

        self.hidden_size = hidden_size
        self.output_size = output_size

        self.out = nn.Linear(hidden_size, output_size)
        self.sm = nn.LogSoftmax(dim=1)

    def forward(self, inputs):
        assert inputs.size(1) == self.hidden_size
        return self.sm(self.out(inputs))
```

```
import torch.nn.functional as F

class MultiAttention(nn.Module):
    def __init__(self, num_features, hidden_dim, output_dim = 1):
        super(MultiAttention, self).__init__()
        self.num_features = num_features
        self.hidden_dim = hidden_dim
        self.output_dim = output_dim

        self.W_a = nn.Linear(self.num_features, self.hidden_dim)
        self.U_a = nn.Linear(self.hidden_dim, self.hidden_dim)
        self.v_a = nn.Linear(self.hidden_dim, self.output_dim)
                
    def forward(self, features, hidden):

        hidden = hidden.unsqueeze(1)
        atten_1 = self.W_a(features)
        atten_2 = self.U_a(hidden)

        atten_tan = torch.tanh(atten_1+atten_2)
        atten_score = self.v_a(atten_tan)
        atten_weight = F.softmax(atten_score, dim = 1)

        context = torch.sum(atten_weight * features,  dim = 1)
        atten_weight = atten_weight.squeeze(dim=2)
        
        return atten_weight, context
```

```
from torch.autograd import Variable


class DecoderRNN(nn.Module):
    def __init__(self, embedding_dim, hidden_size, output_size, max_length, n_layers=3, 
                 dropout=0.3, use_attention=True):
        super(DecoderRNN, self).__init__()

        self.embedding_dim = embedding_dim
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.n_layers = n_layers
        self.dropout = dropout
        self.max_length = max_length
        self.use_attention = use_attention

        self.embedding = nn.Embedding(output_size, embedding_dim)
        
        if self.use_attention:
            self.attn1 = nn.Linear(embedding_dim, hidden_size)
            self.attn2 = nn.Linear(1, max_length)
            self.attn3 = nn.Linear(hidden_size, 1)
            self.attn_sm = nn.Softmax(dim=1)
            self.attn_ctx = nn.Linear(hidden_size + hidden_size, hidden_size)
            self.attn_tanh = nn.Tanh()
        
        self.rnn = nn.LSTM(embedding_dim, hidden_size, n_layers,
                           dropout=dropout, batch_first=True)
        self.generator = Generator(hidden_size, output_size)
        
    def step(self, batch_input, hidden, encoder_output):
        # hidden: (n_layers x B x N, n_layers x B x N)
        # encoder_output: B x L x N
        # h_t: B x N x 1
        # embedded:  B x E
        # attn_weights: B x 1 x L
        # context: B x 1 x N
        # rnn_input: B x N
        
        embedded = self.embedding(batch_input)
        _, hidden = self.rnn(embedded.unsqueeze(1), hidden)
        if self.use_attention:
            h_t = hidden[0][-1].unsqueeze(2)
            mapped_encoder_outputs = self.attn1(encoder_output)
            mapped_decoder_outputs = self.attn2(h_t).transpose(1, 2)
  
            attn_weights = self.attn_sm(self.attn3(torch.tanh(mapped_encoder_outputs + mapped_decoder_outputs))).transpose(1, 2)
            max_length = encoder_output.size(1)
            context = torch.bmm(attn_weights[:, :, :max_length], encoder_output)
            output = self.attn_tanh(self.attn_ctx(torch.cat((context, h_t.transpose(1, 2)), 2)))
        return output, hidden

    def init_state(self, batch_size, bos_index):
        initial_input = Variable(torch.zeros((batch_size,)).type(torch.LongTensor), requires_grad=False)
        initial_input = torch.add(initial_input, bos_index)
        initial_input = initial_input.cuda() if next(self.parameters()).is_cuda else initial_input
        return initial_input

    def forward(self, current_input, hidden, length, encoder_output, target_ids=None):
        batch_size = current_input.size(0)
        outputs = Variable(torch.zeros(batch_size, length, self.output_size), requires_grad=False)
        outputs = outputs.cuda() if next(self.parameters()).is_cuda else outputs

        for t in range(length):
            output, hidden = self.step(current_input, hidden, encoder_output)
            scores = self.generator.forward(output.squeeze(1))
            outputs[:, t, :] = scores
            if target_ids is None:
                top_indices = scores.topk(1, dim=1)[1].view(-1)
                current_input = top_indices
            else:
                current_input = target_ids[:, t]
        return outputs, hidden
```

```
from torch.optim import Adam
from pytorch_lightning import LightningModule

class Seq2Seq(LightningModule):
    def __init__(self, input_vocab_size, output_vocab_size,
                 embedding_dim=100, rnn_size=100,
                 encoder_n_layers=2, decoder_n_layers=2,
                 dropout=0.3, max_length=20,
                 bidirectional=True, use_attention=True):
        super(Seq2Seq, self).__init__()

        self.embedding_dim = embedding_dim
        self.input_vocab_size = input_vocab_size
        self.output_vocab_size = output_vocab_size
        self.rnn_size = rnn_size
        self.encoder_n_layers = encoder_n_layers
        self.decoder_n_layers = decoder_n_layers
        self.dropout = dropout
        self.max_length = max_length
        self.bidirectional = bidirectional
        self.use_attention = use_attention
        self.criterion = nn.NLLLoss()

        self.encoder = EncoderRNN(self.input_vocab_size, embedding_dim, rnn_size, dropout=dropout,
                                  n_layers=encoder_n_layers, bidirectional=bidirectional)
        self.decoder = DecoderRNN(embedding_dim, rnn_size, self.output_vocab_size, dropout=dropout,
                                  max_length=max_length, n_layers=decoder_n_layers, 
                                  use_attention=use_attention)

    def forward(self, source_ids, target_ids=None, bos_index=1):
        batch_size = source_ids.size(0)
        encoder_output, encoder_hidden = self.encoder.forward(source_ids)
        current_input = self.decoder.init_state(batch_size, bos_index)
        scores, _ = self.decoder.forward(
            current_input, encoder_hidden, self.max_length,
            encoder_output, target_ids
        )
        loss = None
        if target_ids is not None:
            loss = self.criterion(scores.transpose(1, 2), target_ids)
        return loss, scores
    
    def training_step(self, batch, batch_nb):
        train_loss, _ = self(**batch)
        self.log("train_loss", train_loss, prog_bar=True)
        return train_loss

    def validation_step(self, batch, batch_nb):
        val_loss, _ = self(**batch)
        self.log("val_loss", val_loss, prog_bar=True, logger=True)
    
    def configure_optimizers(self):
        optimizer = Adam(self.parameters(), lr=0.001)
        return [optimizer]
```

```
from torch.autograd import Variable


class DecoderRNN(nn.Module):
    def __init__(self, embedding_dim, hidden_size, output_size, max_length, n_layers=3, 
                 dropout=0.3, use_attention=True):
        super(DecoderRNN, self).__init__()

        self.embedding_dim = embedding_dim
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.n_layers = n_layers
        self.dropout = dropout
        self.max_length = max_length
        self.use_attention = use_attention

        self.embedding = nn.Embedding(output_size, embedding_dim)
        
        if self.use_attention:
            #self.attn1 = nn.Linear(embedding_dim, hidden_size)
            self.attn_sm = nn.Softmax(dim=1)
            self.attn_ctx = nn.Linear(hidden_size + hidden_size, hidden_size)
            self.attn_tanh = nn.Tanh()

        
        self.rnn = nn.LSTM(embedding_dim, hidden_size, n_layers,
                           dropout=dropout, batch_first=True)
        self.generator = Generator(hidden_size, output_size)
        
    def step(self, batch_input, hidden, encoder_output):
        # hidden: (n_layers x B x N, n_layers x B x N)
        # encoder_output: B x L x N
        # h_t: B x N x 1
        # embedded:  B x E
        # attn_weights: B x 1 x L
        # context: B x 1 x N
        # rnn_input: B x N
        
        embedded = self.embedding(batch_input)
        _, hidden = self.rnn(embedded.unsqueeze(1), hidden)
        if self.use_attention:
            mapped_encoder_outputs = encoder_output
            h_t = hidden[0][-1].unsqueeze(2)
            H = h_t.size()[1]
            attn_weights = self.attn_sm(torch.bmm(mapped_encoder_outputs, h_t)/H**(1/2)).transpose(1, 2)
            max_length = encoder_output.size(1)
            context = torch.bmm(attn_weights[:, :, :max_length], encoder_output)
            output = self.attn_tanh(self.attn_ctx(torch.cat((context, h_t.transpose(1, 2)), 2)))
        return output, hidden

    def init_state(self, batch_size, bos_index):
        initial_input = Variable(torch.zeros((batch_size,)).type(torch.LongTensor), requires_grad=False)
        initial_input = torch.add(initial_input, bos_index)
        initial_input = initial_input.cuda() if next(self.parameters()).is_cuda else initial_input
        return initial_input

    def forward(self, current_input, hidden, length, encoder_output, target_ids=None):
        batch_size = current_input.size(0)
        outputs = Variable(torch.zeros(batch_size, length, self.output_size), requires_grad=False)
        outputs = outputs.cuda() if next(self.parameters()).is_cuda else outputs

        for t in range(length):
            output, hidden = self.step(current_input, hidden, encoder_output)
            scores = self.generator.forward(output.squeeze(1))
            outputs[:, t, :] = scores
            if target_ids is None:
                top_indices = scores.topk(1, dim=1)[1].view(-1)
                current_input = top_indices
            else:
                current_input = target_ids[:, t]
        return outputs, hidden
```

```
from pyonmttok import Tokenizer

pyonmt_tokenizer = Tokenizer("aggressive", joiner_annotate=False)

def tokenize(text, lower=True):
    text = str(text).strip().replace("\n", " ").replace("\xa0", " ")
    if lower:
        text = text.lower()
    tokens, _ = pyonmt_tokenizer.tokenize(text)
    return tokens
```

```
from collections import Counter
from typing import List, Tuple
import os
import torch
from bpemb import BPEmb

class Vocabulary:
    def __init__(self, lang):
        self.index2word = list()
        self.word2index = dict()
        self.word2count = Counter()
        self.bpe_model = BPEmb(lang=lang, dim=50)
        self.reset()

    def get_pad(self):
        return self.word2index["<pad>"]

    def get_bos(self):
        return self.word2index["<bos>"]

    def get_eos(self):
        return self.word2index["<eos>"]

    def get_unk(self):
        return self.word2index["<unk>"]

    def add_sentence(self, sentence):
        for word in tokenize(sentence.strip()):
            self.add_word(word)

    def add_word(self, word):
        word = self.bpe_model.encode(word)
        if len(word) > 0:
          if (word[0][:1] == "▁"):
            word[0] = word[0][1:]

        for bpe_char in word:
          if bpe_char not in self.word2index:
              self.word2index[bpe_char] = len(self.index2word)
              self.word2count[bpe_char] += 1
              self.index2word.append(bpe_char)
          else:
              self.word2count[bpe_char] += 1
    
    def has_word(self, word) -> bool:
        return word in self.word2index

    def add_file(self, filename: str):
        with open(filename, "r", encoding="utf-8") as r:
            for line in r:
                self.add_sentence(line)

    def get_index(self, word):
        if word in self.word2index:
            return self.word2index[word]
        return self.get_unk()

    def get_word(self, index):
        return self.index2word[index]

    def size(self):
        return len(self.index2word)

    def is_empty(self):
        empty_size = 4
        return self.size() <= empty_size

    def shrink(self, n):
        best_words = self.word2count.most_common(n)
        self.reset()
        for word, count in best_words:
            self.add_word(word)
            self.word2count[word] = count

    def reset(self):
        self.word2count = Counter()
        self.index2word = ["<pad>", "<bos>", "<eos>", "<unk>"]
        self.word2index = {word: index for index, word in enumerate(self.index2word)}

    def tokenize(self, sentence: str, max_length: int) -> List[int]:
        words = tokenize(sentence)
        
        subwords = []
        for word in words:
          word = self.bpe_model.encode(word)
          if len(word) > 0:
            if (word[0][:1] == "▁"):
              word[0] = word[0][1:]
          subwords += word
        
        indices = [self.get_index(subword) for subword in subwords]
        indices = indices[:max_length-2]
        indices.insert(0, self.get_bos())
        indices.append(self.get_eos())
        mask = [1 for _ in range(len(indices))]
        indices += [self.get_pad() for _ in range(max_length - len(indices))]
        mask += [0 for _ in range(max_length - len(indices))]
        return torch.tensor(indices), torch.tensor(mask)
```

```
import torch
from torch.utils.data import Dataset

class TranslationVanillaDataset(Dataset):
    def __init__(
        self,
        source_file,
        target_file,
        source_vocabulary,
        target_vocabulary,
        source_max_tokens_count=20,
        target_max_tokens_count=20
    ):
        self.pairs = []
        with open(source_file, "r") as source_file, open(target_file, "r") as target_file:
            for source, target in zip(source_file, target_file):
                self.pairs.append((source, target))
        self.source_vocabulary = source_vocabulary
        self.target_vocabulary = target_vocabulary
        self.source_max_tokens_count = source_max_tokens_count
        self.target_max_tokens_count = target_max_tokens_count
    
    def __len__(self):
        return len(self.pairs)

    def __getitem__(self, index):
        source, target = self.pairs[index]
        source_ids, _ = self.source_vocabulary.tokenize(source, max_length=self.source_max_tokens_count)
        target_ids, _ = self.target_vocabulary.tokenize(target, max_length=self.target_max_tokens_count)
        return {
            "source_ids": source_ids,
            "target_ids": target_ids
        }
```

```
from torch.utils.data import DataLoader

BATCH_SIZE = 64

train_dataset = TranslationVanillaDataset(TRAIN_SOURCE_FILE, TRAIN_TARGET_FILE, ru_vocab, en_vocab)
val_dataset = TranslationVanillaDataset(VAL_SOURCE_FILE, VAL_TARGET_FILE, ru_vocab, en_vocab)
test_dataset = TranslationVanillaDataset(TEST_SOURCE_FILE, TEST_TARGET_FILE, ru_vocab, en_vocab)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE)
```

```
seq2seq_model = Seq2Seq(ru_vocab.size(), en_vocab.size())
```

```
from pytorch_lightning import Trainer as LightningTrainer
from pytorch_lightning.callbacks import EarlyStopping

early_stop_callback = EarlyStopping(
    monitor="val_loss",
    min_delta=0.0,
    patience=4,
    verbose=True,
    mode="min"
)

ltrainer = LightningTrainer(
    gpus=1,
    accumulate_grad_batches=1,
    max_epochs=5,
    val_check_interval=0.2,
    gradient_clip_val=5.0,
    logger=False,
    callbacks=[early_stop_callback]
)
ltrainer.fit(seq2seq_model, train_loader, val_loader)
```

```
class Translator:
    def __init__(
        self,
        model: Seq2Seq,
        src_vocabulary: Vocabulary,
        tgt_vocabulary: Vocabulary, 
        use_cuda: bool
    ):
        self.model = model  # type: Seq2Seq
        self.src_vocabulary = src_vocabulary  # type: Vocabulary
        self.tgt_vocabulary = tgt_vocabulary  # type: Vocabulary
        self.use_cuda = use_cuda

    def translate_sentence(self, sentence: str):
        source_ids = self.sentence_to_tensor(sentence, self.src_vocabulary)
        outputs = self.translate(source_ids)

        translated = list(outputs.cpu().detach().numpy())
        words = []
        for i in translated:
            word = self.tgt_vocabulary.get_word(i)
            if word == "<bos>":
                continue
            if word == "<eos>" or word == "<pad>":
                break
            words.append(word)
        return " ".join(words)

    def translate(self, inputs):
        self.model.eval()
        inputs = inputs.to("cuda")
        batch_size = inputs.size(0)
        bos_index = self.tgt_vocabulary.get_bos()
        _, scores = self.model.forward(inputs, bos_index=bos_index)
        max_length = scores.size(1)
      
        output_variable = Variable(torch.zeros(batch_size, max_length).type(torch.LongTensor))
        output_variable = output_variable.cuda() if self.use_cuda else output_variable
        for t in range(max_length):
            output_variable[:, t] = scores[:, t, :].topk(1, dim=1)[1].view(-1)
        output_variable = output_variable.detach()
        return output_variable[0]

    def sentence_to_tensor(self, sentence, vocabulary):
        indices, _ = vocabulary.tokenize(sentence, max_length=self.model.max_length)
        tensor = indices.unsqueeze(0)
        tensor = tensor.cuda() if self.use_cuda else tensor
        return tensor
```

```
from tqdm import tqdm
from nltk.translate.bleu_score import corpus_bleu

def evaluate_bleu(dataset, nrows=None):
    references = []
    hypotheses = []
    n = len(dataset) if nrows is None else min(len(dataset), nrows)
    for i in tqdm(range(n)):
        source, target = dataset.pairs[i]
        prediction = translator.translate_sentence(source)
        references.append([" ".join(tokenize(target))])
        hypotheses.append(" ".join(tokenize(prediction)))
    return corpus_bleu(references, hypotheses)

score = evaluate_bleu(test_dataset, 100)
print()
print("BLEU: {:.2f}".format(score * 100.0))
```

```
import numpy as np

def search(model, src_input, k=1, sequence_max_len=25, bos_index = 1):

    k_beam = [(0, [0]*(sequence_max_len+1))]

    for l in range(sequence_max_len):
        all_k_beams = []
        for prob, sent_predict in k_beam:
            _, predicted = model.forward(src_input,  bos_index = bos_index)
            predicted = torch.softmax(predicted, dim=2)

            predicted = predicted[0].detach().numpy()
            possible_k = predicted[l].argsort()[-k:][::-1]

            probs = 0
            for i in range(l):
              if predicted[i][sent_predict[i+1]] > 0:
                probs += np.log(predicted[i][sent_predict[i+1]])

            for next_wid in possible_k:
              if predicted[l][next_wid] > 0:
                probs += np.log(predicted[l][next_wid])

                all_k_beams += [(probs, list(sent_predict[:l+1])+[next_wid]+[0]*(sequence_max_len-l-1))]

        k_beam = sorted(all_k_beams)[-k:]

    return k_beam
```

```
class Translator:
    def __init__(
        self,
        model: Seq2Seq,
        src_vocabulary: Vocabulary,
        tgt_vocabulary: Vocabulary, 
        use_cuda: bool
    ):
        self.model = model  # type: Seq2Seq
        self.src_vocabulary = src_vocabulary  # type: Vocabulary
        self.tgt_vocabulary = tgt_vocabulary  # type: Vocabulary
        self.use_cuda = use_cuda

    def translate_sentence(self, sentence: str):
        source_ids = self.sentence_to_tensor(sentence, self.src_vocabulary)
        outputs = self.translate_beam_search(source_ids)

        translated = list(outputs.cpu().detach().numpy())
        words = []
        for i in translated:
            word = self.tgt_vocabulary.get_word(i)
            if word == "<bos>":
                continue
            if word == "<eos>" or word == "<pad>":
                break
            words.append(word)
        return " ".join(words)

    def translate(self, inputs):
        self.model.eval()
        inputs = inputs
        batch_size = inputs.size(0)
        bos_index = self.tgt_vocabulary.get_bos()
        
        _, scores = self.model.forward(inputs, bos_index=bos_index)
        max_length = scores.size(1)
      
        output_variable = Variable(torch.zeros(batch_size, max_length).type(torch.LongTensor))
        output_variable = output_variable.cuda() if self.use_cuda else output_variable
        for t in range(max_length):
            output_variable[:, t] = scores[:, t, :].topk(1, dim=1)[1].view(-1)
        output_variable = output_variable.detach()
        return output_variable[0]

    def translate_beam_search(self, inputs):
        self.model.eval()
        inputs = inputs
        batch_size = inputs.size(0)
        bos_index = self.tgt_vocabulary.get_bos()
        _, scores = self.model.forward(inputs, bos_index=bos_index)
        beam_width = 5

        max_length = scores.size(1)
        k_beams = search(self.model, inputs, k=beam_width, sequence_max_len=max_length, bos_index = bos_index)

        return torch.LongTensor(k_beams[-1][1][1:])

    def sentence_to_tensor(self, sentence, vocabulary):
        indices, _ = vocabulary.tokenize(sentence, max_length=self.model.max_length)
        tensor = indices.unsqueeze(0)
        tensor = tensor.cuda() if self.use_cuda else tensor
        return tensor
```
